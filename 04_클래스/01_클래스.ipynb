```python
class Calculator:
    def __init__(self):
        self.result = 0
    
    def add(self, num):
        self.result = self.result + num   # self.result += num
        return self.result 
    
cal1 = Calculator()
cal2 = Calculator()

print(cal1.add(3))
print(cal1.add(4))

print(cal2.add(5))
print(cal2.add(10))
```

```python
# 곱하기, 빼기, 나누기 기능 만들기
## mul() - 곱하기 
## sub() - 빼기
## div() - 나누기 

### 계산기 2개를 만들어서 (cal1, cal2)
### cal1(10, 20)
### cal2(20, 5)


class FourCal:
    def setdata(self, first, second): 
        self.first = first
        self.second = second
        
    
    def add(self):
        result = self.first + self.second
        return result
    
    def mul(self): 
        result = self.first * self.second
        return result

    def sub(self): 
        result = self.first - self.second
        return result
    
    def div(self): 
        result = self.first / self.second
        return result
    
cal1 = FourCal()
cal1.setdata(10, 20)
cal1.add()
```

```python
class MoreFourCal(FourCal):
    def pow(self): 
        result = self.first ** self.second
        return result 
    

cal4 = MoreFourCal()
cal4.setdata(2, 4) 
cal4.pow()
```

```python
# 매서드 오버라이딩

class SafeFourCal(FourCal): 
    def div(self):
        if self.second == 0: 
            return 0 
        
        else: 
            return self.first / self.second
        

cal1 = FourCal()
cal1.setdata(4, 0)
# cal1.div()

cal2 = SafeFourCal()
cal2.setdata(4, 0)
cal2.div()
```
